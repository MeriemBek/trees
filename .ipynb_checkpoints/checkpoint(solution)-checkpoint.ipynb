```python
def plot_correlation_map( df ):

    corr = df.corr()
    s , ax = plt.subplots( figsize =( 12 , 10 ) ) 
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True ) 
    s = sns.heatmap( 
        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 

        annot = True, 
        annot_kws = { 'fontsize' : 12 }

        )
# This function draws a heatmap to show how much the dataframe's columns are correlated:
#1- It calculates the correlation between all columns in a dataframe .
#2- Create figure and a set of sub plot as a start point.
#3- Choose the pallette's color .
#4- Draw a heatmap ,where the data is the correlation we calculated in the first step and change
    #all numerical values to colored spaces (it also changes colorbar size to look smaller,and set up the font size)
```
